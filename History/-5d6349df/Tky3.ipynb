# style class

```python
np.random.seed(0)
df2 = pd.DataFrame(np.random.randn(10,4), columns=['A','B','C','D'])
df2
df2.style
```

```python
def style_negative(v, props=''):
    return props if v < 0 else None
def highlight_max(s, props=''):
    return np.where(s == np.nanmax(s.values), props, '')

s2 = (df2.style
        .applymap(style_negative, props='color:red;')
        .applymap(lambda v: 'opacity: 60%;' if (v < 0.3) and (v > -0.3) else None)
        .set_caption("my captions")
        # .format('{:.4f}', escape='latex')
        .format('{:.4f}')
        # .apply(highlight_max, props='color:white;background-color:blue', axis=0)
        # .to_latex()
    )

s2
# s2.to_html("hello.html")

# print(s2.to_latex(convert_css=True, hrules=True))
print(s2.to_latex(  convert_css=True, hrules=True, position_float='centering', 
                    clines='all;data',
                    sparse_index=True, sparse_columns=True, multirow_align=True, multicol_align=True
))
```
