# barplot 
- that is, show the value counts of a feature

```python
# pd
tips.day.value_counts().plot.bar().set(
    title='bar graph', yticks=range(10,100,10), xlabel='days', ylabel='counts')
plt.xticks(rotation=0)

# plt
plt.figure()
plt.bar(tips.day.value_counts().index, tips.day.value_counts().values, width=0.5)

# sns 
plt.figure()
sns.countplot(data=tips, x='day') # do not need to specify data in this case 
sns.countplot(x=tips.day, figure=plt.figure(), order=tips.day.value_counts().index)

pass;
```

```python
# countplot by group 

# pd
tips.groupby(['day','sex']).size().unstack().plot.bar() # figure level 
tips.pivot_table(index='day', columns='sex', aggfunc='size').plot.bar()

# sns
sns.countplot(data=tips, x='day', hue='sex', figure=plt.figure()) # axis level...
pass;
```

# barplots

```python
# ---------------------------------- 1d case --------------------------------- #
# pd 
tips.pivot_table(index='day', values='tip').plot.bar()

# sns
sns.barplot(data=tips, x='day', y='tip', ci=None, figure=plt.figure())

# ---------------------------------- 2d case --------------------------------- #
# pd
tips.pivot_table(index='day', columns='sex', values='tip').plot.bar()

# sns
sns.barplot(data=tips, x='day', y='tip', hue='sex', ci=None, figure=plt.figure())

# ---------------------------------- 3d case --------------------------------- #
sns.catplot(data=tips, x='day', y='tip', hue='sex', col='smoker', ci=None, kind='bar')

pass;
```

# line plots

```python
# x = discrete variable, y = count of x
# pd
tips.value_counts('size').sort_index().plot()

# sns
temp = tips.value_counts('size')
sns.lineplot(x=temp.index, y=temp.values, ci=None, figure=plt.figure())
pass;
```
