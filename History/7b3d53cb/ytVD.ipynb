```python
attend_files = sorted(glob('/Users/whlin/Library/CloudStorage/OneDrive-HKUSTConnect/Documents/python/data/raymond_attendance_count/processed/[0-9]*.txt'))
n_cols = len(attend_files)

ids = pd.read_excel('/Users/whlin/Library/CloudStorage/OneDrive-HKUSTConnect/Documents/python/data/raymond_attendance_count/WholeClass.xlsx', usecols='B')['ID']

# first, build a df to store the final results
final_df = pd.DataFrame(np.zeros([len(ids), n_cols], dtype=int), index=ids, columns=[f"A{i}" for i in range(1, n_cols+1)])

final_df

for file in attend_files:
    num = pathlib.Path(file).stem
    with open(file, 'r') as f:
        for sid in f: # each line is a sid
            sid_strip = int(sid.strip())
            if sid_strip in final_df.index: # some students might have dropped...
                final_df.loc[sid_strip, f"A{num}"] = 1

final_df    
final_df.to_excel("~/Downloads/hello.xlsx")
```
