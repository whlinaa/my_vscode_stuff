```python
# multiIndex access example
tips_agg = tips.groupby(['sex','day'])[['total_bill','tip','size']].agg(['mean','median'])
print(tips_agg)
#             total_bill               tip             size       
#                   mean  median      mean median      mean median
# sex    day                                                      
# Male   Thur  18.714667  16.975  2.980333  2.530  2.433333    2.0
#        Fri   19.857000  17.215  2.693000  2.600  2.100000    2.0
#        Sat   20.802542  18.240  3.083898  3.000  2.644068    2.0
#        Sun   21.887241  20.725  3.220345  3.085  2.810345    2.0
# Female Thur  16.715312  13.785  2.575625  2.005  2.468750    2.0
#        Fri   14.145556  15.380  2.781111  3.000  2.111111    2.0
#        Sat   19.680357  18.360  2.801786  2.625  2.250000    2.0
#        Sun   19.872222  17.410  3.367222  3.500  2.944444    3.0

# [] accessor. we can use it to access the outer level of column index only..
print(tips_agg[['total_bill','size']][:3])
#           total_bill              size       
#                 mean  median      mean median
# sex  day                                     
# Male Thur  18.714667  16.975  2.433333    2.0
#      Fri   19.857000  17.215  2.100000    2.0
#      Sat   20.802542  18.240  2.644068    2.0
print(tips_agg[tips_agg.loc[:, ('total_bill','mean')]>20][:3])
#          total_bill               tip             size       
#                mean  median      mean median      mean median
# sex  day                                                     
# Male Sat  20.802542  18.240  3.083898  3.000  2.644068    2.0
#      Sun  21.887241  20.725  3.220345  3.085  2.810345    2.0
#           total_bill               tip             size       

# slicing works as usual 
print(tips_agg[:3])
#                 mean  median      mean median      mean median
# sex  day                                                      
# Male Thur  18.714667  16.975  2.980333   2.53  2.433333    2.0
#      Fri   19.857000  17.215  2.693000   2.60  2.100000    2.0
#      Sat   20.802542  18.240  3.083898   3.00  2.644068    2.0
#      total_bill               tip             size       

print(tips_agg.loc['Male'][:3]) # index, outer level 
#            mean  median      mean median      mean median
# day                                                      
# Thur  18.714667  16.975  2.980333   2.53  2.433333    2.0
# Fri   19.857000  17.215  2.693000   2.60  2.100000    2.0
# Sat   20.802542  18.240  3.083898   3.00  2.644068    2.0
# tips_agg.loc['Thur'][:3] # not work 

print(tips_agg.loc[('Male','Thur')] [:3]) # index, outer and inner level 
# total_bill  mean      18.714667
#             median    16.975000
# tip         mean       2.980333
# Name: (Male, Thur), dtype: float64
# tips_agg.loc['Male','Thur'][:3] # although without parenthesis works, don't use it!!

print(tips_agg.loc[('Male','Thur'), 'tip'] [:3]) # index, outer and inner level, columns, outer level
# mean      2.980333
# median    2.530000
# Name: (Male, Thur), dtype: float64

print(tips_agg.loc[('Male','Thur'), (['tip','size'])][:3]) # index, outer and inner level; columns, outer level
# tip   mean      2.980333
#       median    2.530000
# size  mean      2.433333
# Name: (Male, Thur), dtype: float64

print(tips_agg.loc[('Male','Thur'), (['tip','size'], 'mean')][:3]) # index, outer and inner level; columns, outer and inner level
# tip   mean    2.980333
# size  mean    2.433333
# Name: (Male, Thur), dtype: float64

print(tips_agg.loc[(['Male','Female'],['Thur','Sun']), (['tip','size'], ['mean','median'])][:3]) # index, outer and inner level; columns, outer and inner level
#                   tip             size       
#                  mean median      mean median
# sex    day                                   
# Male   Thur  2.980333  2.530  2.433333    2.0
#        Sun   3.220345  3.085  2.810345    2.0
# Female Thur  2.575625  2.005  2.468750    2.0
```

```python
df1 = pd.DataFrame({'key': ['a', 'a', 'b', 'c'], 'data1': range(4)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'], 'data2': range(3)})
# df1.merge(df2) # `how` defaults to `inner`. Merge by the 'key' column. Note that I didnt specify which column to join on. If that information is not specified, merge uses the overlapping column names as the keys. Its a good practice to specify explicitly, though.

# in below, `on='key'` is implied
print(df1.merge(df2, how='inner')) # fewer records
#   key  data1  data2
# 0   a      0      0
# 1   a      1      0
# 2   b      2      1
print(df1.merge(df2, how='left')) # more than inner
#   key  data1  data2
# 0   a      0    0.0
# 1   a      1    0.0
# 2   b      2    1.0
# 3   c      3    NaN
print(df1.merge(df2, how='right')) # more than inner
#   key  data1  data2
# 0   a    0.0      0
# 1   a    1.0      0
# 2   b    2.0      1
# 3   d    NaN      2
print(df1.merge(df2, how='outer')) # more than inner, left, right 
#   key  data1  data2
# 0   a    0.0    0.0
# 1   a    1.0    0.0
# 2   b    2.0    1.0
# 3   c    3.0    NaN
# 4   d    NaN    2.0
print(df1.merge(df2, how='cross')) # cartesian product. should have `len(df1)*len(df2)` records
#    key_x  data1 key_y  data2
# 0      a      0     a      0
# 1      a      0     b      1
# 2      a      0     d      2
# 3      a      1     a      0
# 4      a      1     b      1
# 5      a      1     d      2
# 6      b      2     a      0
# 7      b      2     b      1
# 8      b      2     d      2
# 9      c      3     a      0
# 10     c      3     b      1
# 11     c      3     d      2
```

```python
df3 = pd.DataFrame({'lkey': ['a', 'a', 'b', 'c'], 'data1': range(4)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'], 'data2': range(3)})

print(pd.merge(df3, df4, left_on='lkey', right_on='rkey', how='inner')) # If the column names are different in each object, you can specify them separately. You may notice that the 'c' and 'd' values and associated data are missing from the result. By default merge does an 'inner' join; the keys in the result are the intersection, or the common set found in both tables. Other possible options are 'left', 'right', and 'outer'. The outer join takes the union of the keys, combining the effect of applying both left and right joins:
#   lkey  data1 rkey  data2
# 0    a      0    a      0
# 1    a      1    a      0
# 2    b      2    b      1
print(pd.merge(df3, df4, left_on='lkey', right_on='rkey', how='left'))
#   lkey  data1 rkey  data2
# 0    a      0    a    0.0
# 1    a      1    a    0.0
# 2    b      2    b    1.0
# 3    c      3  NaN    NaN
print(pd.merge(df3, df4, left_on='lkey', right_on='rkey', how='right'))
#   lkey  data1 rkey  data2
# 0    a    0.0    a      0
# 1    a    1.0    a      0
# 2    b    2.0    b      1
# 3  NaN    NaN    d      2
print(pd.merge(df3, df4, left_on='lkey', right_on='rkey', how='outer'))
#   lkey  data1 rkey  data2
# 0    a    0.0    a    0.0
# 1    a    1.0    a    0.0
# 2    b    2.0    b    1.0
# 3    c    3.0  NaN    NaN
# 4  NaN    NaN    d    2.0
# print(pd.merge(df3, df4, left_on='lkey', right_on='rkey', how='cross'))
```
