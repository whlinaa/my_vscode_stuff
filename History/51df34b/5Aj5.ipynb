# MultiIndex 

## how to create multiIndex

```python
# create multiindex by `from_frame`
df = pd.DataFrame(
    [["bar", "one"], ["bar", "two"], ["foo", "one"], ["foo", "two"]],
    columns=["first", "second"],
)
df
pd.MultiIndex.from_frame(df)
```

```python
# use `from_tuples`
arrays = [
    ["bar", "bar", "baz", "baz", "foo", "foo", "qux", "qux"],
    ["one", "two", "one", "two", "one", "two", "one", "two"],
]
tuples = list(zip(*arrays))
tuples
idx = pd.MultiIndex.from_tuples(tuples, names=["first", "second"])
# len(idx)
df = pd.DataFrame(np.random.randn(3, len(idx)), index=crange(0,3, lower=False), columns=idx)
df
```

```python
tuples = [(1,1),(1,2),(2,1),(2,2)] # 4 rows
products = [[1,2,3], [4,5,6]] # 3*3 = 9 row
arrays = [[1, 1, 2, 2], ['red', 'blue', 'red', 'blue']] # 4 rows

print("from_product")
idx = pd.MultiIndex.from_product(products)
idx

print("from_tuples")
pd.MultiIndex.from_tuples(tuples)

print("from_arrays")
pd.MultiIndex.from_arrays(arrays)

print("from_frame")
pd.MultiIndex.from_frame(pd.DataFrame({'A':[1,2,3,4,5],'B':[6,7,8,9,10]}))

idx.get_level_values(0)
idx.get_level_values(1)
```

```python
# create a df with multiindex directly
df = pd.DataFrame([[1,2,3],[4,5,6]],  
                    index=[[-1,-2],[-3,-4]], 
                    columns=[['a','a','c'],['d','e','f']]
                    ).rename_axis(index=['i1','i2'], columns=['c1','c2'])
df
```

# merge and concat

```python
df1 = pd.DataFrame({'key': ['a', 'a', 'b', 'c'], 'data1': range(4)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'], 'data2': range(3)})
# df1.merge(df2) # `how` defaults to `inner`. Merge by the 'key' column. Note that I didn’t specify which column to join on. If that information is not specified, merge uses the overlapping column names as the keys. It’s a good practice to specify explicitly, though.

# in below, `on='key'` is implied
print(df1.merge(df2, how='inner')) # fewer records
#   key  data1  data2
# 0   a      0      0
# 1   a      1      0
# 2   b      2      1
print(df1.merge(df2, how='left')) # more than inner
#   key  data1  data2
# 0   a      0    0.0
# 1   a      1    0.0
# 2   b      2    1.0
# 3   c      3    NaN
print(df1.merge(df2, how='right')) # more than inner
#   key  data1  data2
# 0   a    0.0      0
# 1   a    1.0      0
# 2   b    2.0      1
# 3   d    NaN      2
print(df1.merge(df2, how='outer')) # more than inner, left, right 
#   key  data1  data2
# 0   a    0.0    0.0
# 1   a    1.0    0.0
# 2   b    2.0    1.0
# 3   c    3.0    NaN
# 4   d    NaN    2.0
print(df1.merge(df2, how='cross')) # cartesian product. should have `len(df1)*len(df2)` records
#    key_x  data1 key_y  data2
# 0      a      0     a      0
# 1      a      0     b      1
# 2      a      0     d      2
# 3      a      1     a      0
# 4      a      1     b      1
# 5      a      1     d      2
# 6      b      2     a      0
# 7      b      2     b      1
# 8      b      2     d      2
# 9      c      3     a      0
# 10     c      3     b      1
# 11     c      3     d      2
```

```python
df3 = pd.DataFrame({'lkey': ['a', 'a', 'b', 'c'], 'data1': range(4)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'], 'data2': range(3)})

print(pd.merge(df3, df4, left_on='lkey', right_on='rkey', how='inner')) # If the column names are different in each object, you can specify them separately. You may notice that the 'c' and 'd' values and associated data are missing from the result. By default merge does an 'inner' join; the keys in the result are the intersection, or the common set found in both tables. Other possible options are 'left', 'right', and 'outer'. The outer join takes the union of the keys, combining the effect of applying both left and right joins:
#   lkey  data1 rkey  data2
# 0    a      0    a      0
# 1    a      1    a      0
# 2    b      2    b      1
print(pd.merge(df3, df4, left_on='lkey', right_on='rkey', how='left'))
#   lkey  data1 rkey  data2
# 0    a      0    a    0.0
# 1    a      1    a    0.0
# 2    b      2    b    1.0
# 3    c      3  NaN    NaN
print(pd.merge(df3, df4, left_on='lkey', right_on='rkey', how='right'))
#   lkey  data1 rkey  data2
# 0    a    0.0    a      0
# 1    a    1.0    a      0
# 2    b    2.0    b      1
# 3  NaN    NaN    d      2
print(pd.merge(df3, df4, left_on='lkey', right_on='rkey', how='outer'))
#   lkey  data1 rkey  data2
# 0    a    0.0    a    0.0
# 1    a    1.0    a    0.0
# 2    b    2.0    b    1.0
# 3    c    3.0  NaN    NaN
# 4  NaN    NaN    d    2.0
# print(pd.merge(df3, df4, left_on='lkey', right_on='rkey', how='cross'))
```

# styling of tables
## important methods
- highlight_{max,min,null,between,quantile}
- set_caption()
- {background, text}_gradient
- format(precious=2, na_rep='Missing', thousands=',')
- applymap
- apply

```python
# define multiple styling functions
def rain_condition(v):
    if v < 1.75:
        return "Dry"
    elif v < 2.75:
        return "Rain"
    return "Heavy Rain"

def make_pretty(styler):
    (
        styler.set_caption("Weather Conditions")
                .format(rain_condition)
                .format_index(lambda v: v.strftime("%A"))
                .background_gradient(axis=None, vmin=1, vmax=5, cmap="YlGnBu")
    )
    return styler

def style_negative(v, props='color:red;'):
    return props if v<0 else None

def highlight_max(s, props=''):
    return np.where(s == np.nanmax(s.values), props, '')

def style_opaque(v, props='opacity: 50%;'):
    return 'opacity: 50%;' if abs(v)<=0.3 else None

def highlight_extreme(v, props='background:blue;'):
    return props if v in [0, 'N/A','S','F', np.nan] else None
```
